```python
from pathlib import Path
import os

import numpy as np
import pandas as pd
```

```python
def find_project_root(start: Path | None = None) -> Path:
    """Walk upward until pyproject.toml is found."""
    current = (start or Path.cwd()).resolve()

    for candidate in (current, *current.parents):
        if (candidate / "pyproject.toml").exists():
            return candidate

    raise FileNotFoundError(
        "Could not find project root containing pyproject.toml"
    )


ROOT = find_project_root()
IBM_TRAINING = ROOT / "share_IBM_training"
OPTICAL_DATA = ROOT / "Alignment_AI_coding" / "configs"

ROOT, IBM_TRAINING, OPTICAL_DATA
```

```python
asr_df = pd.read_csv(OPTICAL_DATA / "alignment_scoring_result.csv")
asr_df.head()
```

```python
amp_df = pd.read_csv(IBM_TRAINING / "AMP_summary.csv")
amp_df.head()
```

```python
osc_df = pd.read_csv(IBM_TRAINING / "OSC_summary.csv")
osc_df.head()
```

```python
rank_cols = [col for col in asr_df.columns if 'rank' in col.lower()]
print(len(rank_cols))
rank_cols
```

```python
asr_df[rank_cols].head()
```

```python
asr_cols_set = set(asr_df.columns)
amp_cols_set = set(amp_df.columns)
osc_cols_set = set(osc_df.columns)
```

```python
print("Common columns in ASR and AMP:")

common_cols_asr_amp = asr_cols_set.intersection(amp_cols_set)
print(common_cols_asr_amp)
print("Length:", len(common_cols_asr_amp))

print("Common columns in ASR and OSC:")

common_cols_asr_osc = asr_cols_set.intersection(osc_cols_set)
print(common_cols_asr_osc)
print("Length:", len(common_cols_asr_osc))
```

```python
print("Unique columns in ASR:")
unique_cols_asr = asr_cols_set.difference(amp_cols_set).difference(osc_cols_set)
print(unique_cols_asr)
print("Length:", len(unique_cols_asr))
```

```python
print("AMP columns not in ASR:")
unique_cols_amp = amp_cols_set.difference(asr_cols_set)
print(unique_cols_amp)
print("Length:", len(unique_cols_amp))
```

```python
print("OSC columns not in ASR:")
unique_cols_osc = osc_cols_set.difference(asr_cols_set)
print(unique_cols_osc)
print("Length:", len(unique_cols_osc))
```
